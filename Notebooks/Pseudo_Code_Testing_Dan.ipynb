```python
# Import libraries and dependencies
import pandas as pd
from pathlib import Path
```

```python
# testing pct change functionality
combined_data = Path("../resources/dan/sample_combined_data_dan.csv")
combined_data_df = pd.read_csv(combined_data)

# sample output
combined_data_df.head()
```

```python
# set variable(s) for later 'def' use
groupby_month = True
```

```python
# set 'date' column type to date with year+month or with only year format (for easier groupby date, later)
if groupby_month == True:

    # set 'date' column type to date with year+month only format
    combined_data_df['date'] = pd.to_datetime(combined_data_df['date'], format= '%Y%m', exact=False)

elif groupby_month == False:

    # set 'date' column type to date with year only format
    combined_data_df['date'] = pd.to_datetime(combined_data_df['date'], format= '%Y', exact=False)

# sample output
combined_data_df.dtypes
```

```python
#  create multi-index of combined_date with groupby zip and date, then calculate sum for each new row
combined_by_zipdate = combined_data_df.groupby(['zip','date']).sum()

# sample output
combined_by_zipdate.groupby(level=0).head(2)
```

```python
# Calc pct_chg for each column
combined_pct_chg = combined_by_zipdate.pct_change().dropna()

# sample output
combined_pct_chg.head()
```

```python
# check for bad data
combined_pct_chg.isnull()
```

```python
import numpy as np

# Clean up np.inf or -np.inf values with 0
# combined_pct_chg.replace([np.inf, -np.inf], np.nan)
combined_pct_chg.replace([np.inf, -np.inf], 0, inplace=True)

# sample output
combined_pct_chg.head()

# type(combined_pct_chg.iloc[0, 2])
# csv_data["price_increased_count"] = csv_data["price_increased_count"].fillna("Unknown")
# csv_data
# # Cleanse up price_increased_count pct_chg
# # Cleanse nulls from DataFrame by filling na
# csv_data["customer_no"] = csv_data["customer_no"].fillna("Unknown")
# csv_data
```
